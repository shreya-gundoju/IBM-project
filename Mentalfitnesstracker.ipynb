**PRE PROCESSING**

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
  if df[i].dtype=='object':
    df[i]=l.fit_transform(df[i])
```

```python
print(df.shape)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.30,random_state=36)
```

```python
x.shape
```

```python
x_train.shape, x_test.shape
```

**MODEL TRAINING**

```python
from sklearn .linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print('Intercept: ', lr.intercept_)
print('Coefficients array: ', lr.coef_)
```

```python
y_predict=lr.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse=mean_squared_error(y_train,y_predict)
print('mean_squared_error:',mse)
mae=mean_absolute_error(y_train,y_predict)
print('mean_absolute_error:',mae)
print('root_mean_squared_error:',np.sqrt(mse))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_train,y_predict)
score
```

```python
print(f'Training data r-squared:,{lr.score(x_train, y_train)}')
print(f'Test data r-squared:{lr.score(x_test, y_test)}')
```

```python
y_pred=lr.predict(x_test)
```

**VISUALIZING THE TEST DATA**

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```
